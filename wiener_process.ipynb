```python
import numpy as np
import matplotlib.pyplot as plt

# Set up time and parameters
T = 1.0  # Total time
N = 1000  # Number of time steps
dt = T/N  # Time step size
t = np.linspace(0, T, N+1)  # Time vector

# Parameters for various processes
mu = 0.1  # Drift coefficient for generalized processes
sigma = 0.2  # Volatility coefficient for generalized processes

# Wiener Process (Brownian Motion)
W = np.random.normal(0, np.sqrt(dt), N+1).cumsum()

# Generalized Wiener Process
X = mu*t + sigma*W

# Itô Process (we'll use the same generalized form with constant coefficients for simplicity)
Ito = mu*t + sigma*W

# Geometric Brownian Motion (specific case of Itô Process)
S0 = 100  # Initial stock price
GBM = S0*np.exp((mu - 0.5*sigma**2)*t + sigma*W)

# Plotting
fig, axs = plt.subplots(2, 2, figsize=(14, 8))

# Wiener Process
axs[0, 0].plot(t, W)
axs[0, 0].set_title('Wiener Process (Brownian Motion)')

# Generalized Wiener Process
axs[0, 1].plot(t, X)
axs[0, 1].set_title('Generalized Wiener Process')

# Itô Process
axs[1, 0].plot(t, Ito)
axs[1, 0].set_title('Itô Process')

# Geometric Brownian Motion
axs[1, 1].plot(t, GBM)
axs[1, 1].set_title('Geometric Brownian Motion (GBM)')

# Layout adjustments
plt.tight_layout()
plt.show()
```
