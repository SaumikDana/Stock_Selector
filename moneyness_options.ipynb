```python
import matplotlib.pyplot as plt
import numpy as np

# Set seed for reproducibility
np.random.seed(0)

# Generate simulated current market prices
current_prices = np.random.uniform(90, 110, 5)

# Define strike prices according to typical moneyness
strike_prices = np.array([85, 85, 105, 115, 115])  # Preset for ITM, ATM, OTM

# Calculate strike/current price ratio
ratios = strike_prices / current_prices

# Define option types
option_types = ['OTM Puts', 'ITM Calls','ATM Puts/Calls', 'ITM Puts',  'OTM Calls']

# Plot settings
fig, ax = plt.subplots(figsize=(10, 6))
colors = ['blue', 'orange', 'red', 'green', 'purple']
markers = ['o', 's', 'D', '^', 'v']  # Adjusted markers for better visualization

# Plot each option type with its calculated ratio
for i, ratio in enumerate(ratios):
    ax.plot([i + 1], [ratio], color=colors[i], marker=markers[i], markersize=10, label=option_types[i])

# Customize the plot
ax.set_xticks(range(1, len(option_types) + 1))
ax.set_xticklabels(option_types, rotation=45, ha='right')
ax.set_ylabel('Strike/Current Price Ratio')
ax.set_title('Moneyness Represented by Strike/Current Price Ratio')
ax.grid(True, which='both', linestyle='--', linewidth=0.5)
ax.axhline(1, color='black', linewidth=1, linestyle='--')  # Line at ratio=1 to indicate ATM options
ax.legend()

plt.tight_layout()
plt.show()
```
