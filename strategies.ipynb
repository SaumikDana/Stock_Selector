```python
from IPython.display import display
import pandas as pd

# Define the DataFrame styling
def style_dataframes(df):
    return df.style.set_table_styles(
        [{
            'selector': 'th',
            'props': [
                ('font-size', '14pt'),
                ('text-align', 'center'),
                ('background-color', 'lightgray'),
                ('color', 'black')
            ]
        },
        {
            'selector': 'td',
            'props': [
                ('text-align', 'center')
            ]
        }]
    ).hide(axis='index')  # Hiding the index

# Define the bullish strategies DataFrame
df_bullish = pd.DataFrame({
    "Position Type": [
        "Basic Long Call", 
        "Bull Call Spread", 
        "Long Call Ladder",
        "Diagonal Spread (Bullish Variant)", 
        "Ratio Spread (Bullish Variant)", 
        "Naked Short Put"
    ],
    "Market Sentiment": [
        "Bullish", 
        "Moderately Bullish", 
        "Bullish with Higher Breakout Potential",
        "Neutral to Bullish, Time and Price Speculation", 
        "Directionally Bullish with Undefined Risk", 
        "Bullish to Neutral"
    ]
})

# Define the bearish strategies DataFrame
df_bearish = pd.DataFrame({
    "Position Type": [
        "Basic Long Put", 
        "Bear Put Spread", 
        "Long Put Ladder",
        "Diagonal Spread (Bearish Variant)", 
        "Ratio Spread (Bearish Variant)", 
        "Naked Short Call"
    ],
    "Market Sentiment": [
        "Bearish", 
        "Moderately Bearish", 
        "Bearish with Lower Breakout Potential",
        "Neutral to Bearish, Time and Price Speculation", 
        "Directionally Bearish with Undefined Risk", 
        "Bearish to Neutral"
    ]
})

# Define the strategies that can go both ways DataFrame
df_both_ways = pd.DataFrame({
    "Position Type": [
        "Covered Short Call", 
        "Iron Condor", 
        "Long Straddle", 
        "Short Straddle", 
        "Long Strangle", 
        "Short Strangle", 
        "Long Call Butterfly Spread",
        "Long Put Butterfly Spread", 
        "Calendar Spread", 
        "Butterfly Spread with Puts", 
        "Iron Butterfly", 
        "Synthetic Positions"
    ],
    "Market Sentiment": [
        "Mildly Bullish to Neutral", 
        "Neutral, Profits from Low Volatility", 
        "Volatile, Uncertain Direction", 
        "Neutral, Profits from Low Volatility", 
        "Volatile, Uncertain Direction", 
        "Neutral, Profits from Low Volatility", 
        "Neutral to Slightly Bullish", 
        "Neutral to Slightly Bearish", 
        "Neutral, Speculating on Volatility Differences over Time", 
        "Neutral, Profits from Low Volatility", 
        "Neutral, Profits from Low Volatility with Defined Risk", 
        "Depends on the Constructed Payoff Profile"
    ]
})

# Apply the styling and display each DataFrame
styled_df_bullish = style_dataframes(df_bullish)
styled_df_bearish = style_dataframes(df_bearish)
styled_df_both_ways = style_dataframes(df_both_ways)

display(styled_df_bullish)
display(styled_df_bearish)
display(styled_df_both_ways)
```
